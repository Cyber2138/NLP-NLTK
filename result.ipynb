```python
import requests
from bs4 import BeautifulSoup
import os

urls = new
directory = 'C:\\Users\\ADMIN\\Desktop\\sentiment analysis\\dummy_files'

if not os.path.exists(directory):
    os.makedirs(directory)

def extract_and_save_content(url, file_name):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        div_content = soup.find('div', class_='td-post-content')
        if div_content:
            extracted_content = div_content.get_text(strip=True)
            file_path = os.path.join(directory, file_name)
            with open(file_path, 'w', encoding='utf-8') as f:
                f.write(extracted_content)
            print(f"Content saved to {file_name}")
        else:
            print(f"No div with class 'td-post-content' found for {url}")
    else:
        print(f"Failed to retrieve the webpage {url}. Status code: {response.status_code}")

for i, url in enumerate(urls, start=1):
    file_name = f"Blackassign{i:04d}.txt"
    extract_and_save_content(url, file_name)

print("Extraction and saving completed.")
```

```python
import re
from textblob import TextBlob
import textstat

def load_words(file_path):
    """Load words from a file and return a set of words."""
    with open(file_path, 'r', encoding='utf-8') as file:
        words = set(file.read().splitlines())
    return words

def count_syllables(word):
    """Count syllables in a word."""
    return textstat.syllable_count(word)

def count_personal_pronouns(text):
    """Count personal pronouns in the text."""
    pronouns = re.findall(r'\b(I|we|my|ours|us|Our|We|My|Us|I|Ours)\b', text)
    return len(pronouns)

def avg_word_length(words):
    """Calculate average word length."""
    return sum(len(word) for word in words) / len(words) if words else 0

def analyze_text(content):
    """Analyze text and extract metrics."""
    blob = TextBlob(content)
    polarity_score = blob.sentiment.polarity
    subjectivity_score = blob.sentiment.subjectivity

    sentences = textstat.sentence_count(content)
    words = content.split()
    word_count = len(words)
    complex_words = [word for word in words if textstat.syllable_count(word) >= 3]
    complex_word_count = len(complex_words)
    syllables_per_word = sum(count_syllables(word) for word in words) / word_count if word_count else 0
    avg_sentence_length = word_count / sentences if sentences else 0
    percentage_complex_words = (complex_word_count / word_count) * 100 if word_count else 0
    fog_index = textstat.gunning_fog(content)
    avg_number_of_words_per_sentence = avg_sentence_length
    personal_pronouns = count_personal_pronouns(content)
    avg_word_len = avg_word_length(words)

    return {
        'Positive Score': len([word for word in words if TextBlob(word).sentiment.polarity > 0]),
        'Negative Score': len([word for word in words if TextBlob(word).sentiment.polarity < 0]),
        'Polarity Score': polarity_score,
        'Subjectivity Score': subjectivity_score,
        'AVG SENTENCE LENGTH': avg_sentence_length,
        'PERCENTAGE OF COMPLEX WORDS': percentage_complex_words,
        'FOG INDEX': fog_index,
        'AVG NUMBER OF WORDS PER SENTENCE': avg_number_of_words_per_sentence,
        'COMPLEX WORD COUNT': complex_word_count,
        'WORD COUNT': word_count,
        'SYLLABLE PER WORD': syllables_per_word,
        'PERSONAL PRONOUNS': personal_pronouns,
        'AVG WORD LENGTH': avg_word_len
    }

directory_path = 'C:\\Users\\ADMIN\\Desktop\\sentiment analysis\\dummy_files'

entries = os.listdir(directory_path)

results = []

for entry in entries[:100]: 
    full_path = os.path.join(directory_path, entry)
    if os.path.isfile(full_path):
        try:
            with open(full_path, 'r', encoding='utf-8') as file:
                content = file.read()
                metrics = analyze_text(content)
                metrics['Sentiment_rating'] = entry
                results.append(metrics)
        except Exception as e:
            print(f"Could not process file {entry}: {e}")

df = pd.DataFrame(results)

output_file = 'sentiment_analysis_results.csv'
df.to_csv(output_file, index=False)

print(f"Analysis results saved to {output_file}")
```
